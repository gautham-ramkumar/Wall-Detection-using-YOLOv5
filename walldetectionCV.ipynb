```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/home/rgautham20/yolov5/runs/train/exp2/weights/best.pt', force_reload=True)
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('WALL CLASSIFICATION', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
